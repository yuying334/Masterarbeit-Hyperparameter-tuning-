```python
import pandas as pd
import plotly.graph_objs as go
from keras.datasets import *
import os
import sys
import pathlib
from sklearn.manifold import TSNE
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import plotly.graph_objects as go
from sklearn.decomposition import PCA
```

```python
fashion_idx = [
    "T-Shirt",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot",
]
idx_dic={0: 'T-Shirt',
         1: 'Trouser',
         2: 'Pullover',
         3: 'Dress',
         4: 'Coat',
         5: 'Sandal',
         6: 'Shirt',
         7: 'Sneaker',
         8: 'Bag',
         9: 'Ankle boot'}
```

```python
%%time
df_train=pd.read_csv('fashion_mnist_data/fashion-mnist_train.csv')
label_list=[]
df_train['label']=df_train.label.replace(idx_dic)
label_list=df_train['label']
df_train.drop('label', 1, inplace=True)
df_train=df_train.div(255.0)
df_train.insert(0, 'label',label_list, False)
```

```python
print(df_train.shape)
df_train.head()
```

```python
df = df_train.sample(n=10000, random_state=42)
df1 = df.loc[df['label'] != 'T-Shirt', :]
df2 = df1.loc[df['label'] != 'Trouser', :]
df3 = df2.loc[df['label'] != 'Pullover', :]
df4 = df3.loc[df['label'] != 'Coat', :]
df5 = df4.loc[df['label'] != 'Sandal', :]
df6 = df5.loc[df['label'] != 'Shirt', :]
df7 = df6.loc[df['label'] != 'Sneaker', :]
df8 = df7.loc[df['label'] != 'Dress', :]
df8.head(30)
```

```python
x_train = df8.loc[:, df.columns != 'label']
y_train= df8.label.values
print(x_train.iloc[[1]])
```

```python
%%time
pca = PCA(n_components=200)
data_pca = pca.fit_transform(x_train)
```

```python
%%time
tsne = TSNE(n_components=3,verbose=True,perplexity=20.0)
embedding = tsne.fit_transform(data_pca)
```

```python
label_list=[]
label_list=df8['label']
embedding_df = pd.DataFrame(embedding, columns=["x", "y", "z"],index=label_list)
embedding_df.head()
```

```python
%%time
i=0
data=[]
color_list = ['red', 'blue']
groups=embedding_df.groupby("label")
for idx, val in groups:
    scatter = go.Scatter3d(
        name=str(embedding_df.index),
        x=val["x"],
        y=val["y"],
        z=val["z"],
        text=[idx for _ in range(val["x"].shape[0])],
        hoverinfo='text',
        textposition="middle center",
        showlegend=False,
        mode="markers",
        marker=dict(size=3, symbol="circle", color=color_list[i]),
    )
    data.append(scatter)
    i += 1
figure = go.Figure(data)
figure.show()
```
